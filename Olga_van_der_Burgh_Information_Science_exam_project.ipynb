```python
from query_DBpedia import clean, query_DBpedia, convert_XML_to_JSON
from choose_title import create_list_of_titles, choose_title
from metadata_book import get_metadata, get_abstract, get_publication_date, get_author
from metadata_author import get_metadata_author, get_abstract_author
```

```python
if __name__ == "__main__":
    search = ""
    while not search == "q:":
        search = input("Please input your search (or q: to quit): ")
        if search == "q:":
            exit()
        else:
            # 1) Query DBpedia lookup API
            result = query_DBpedia(search)
            # 2) Convert XML response to JSON
            parsed_result = convert_XML_to_JSON(result)
            # 3) Create list of titles
            list_of_titles = create_list_of_titles(parsed_result)
            # 4) Let user choose a title they want to look up
            chosen_title = choose_title(list_of_titles)
            # 5) Get metadata of chosen title
            metadata = get_metadata(chosen_title)
            # 6) Get abstract of chosen title
            abstract = get_abstract(metadata)
            # 7) Get publication date of chosen title
            publication_date = get_publication_date(metadata)
            # 8) Get author of chosen title
            author = get_author(metadata)
            # 9) Get metadata of author
            metadata_author = get_metadata_author(author)
            # 10) Get abstract of author
            abstract_author = get_abstract_author(metadata_author)
```
